There are outlier values on the top, we need to suppress them as below:


```python

Q1 = df['bmi'].quantile(0.25)
Q3 = df['bmi'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df['bmi'] = df['bmi'].apply(lambda x: upper_bound if x > upper_bound else (lower_bound if x < lower_bound else x))

sns.boxplot(df['bmi'])
plt.title('Box plot of BMI')
plt.show()
```

**Scaling**

> Veri Ã¶lÃ§eklendirme, genellikle ÅŸu iki ana nedenden dolayÄ± yapÄ±lÄ±r:

**Makine Ã–ÄŸrenimi AlgoritmalarÄ±nÄ±n PerformansÄ±:** BirÃ§ok makine Ã¶ÄŸrenimi algoritmasÄ±, Ã¶zellikler arasÄ±ndaki bÃ¼yÃ¼klÃ¼k farklarÄ±nÄ± dikkate almaz. Bu nedenle, Ã¶lÃ§eklenmemiÅŸ verilerle Ã§alÄ±ÅŸÄ±rken, bÃ¼yÃ¼k deÄŸerlere sahip Ã¶zellikler diÄŸerlerini baskÄ±layabilir ve modelin yanlÄ±ÅŸ sonuÃ§lar Ã¼retmesine neden olabilir.

**Optimizasyon ve HÄ±zlandÄ±rma:** BazÄ± algoritmalar, verinin Ã¶lÃ§eklenmiÅŸ olmasÄ±nÄ±n modelin eÄŸitimini hÄ±zlandÄ±rabileceÄŸi veya daha iyi sonuÃ§lar verebileceÄŸi durumlar vardÄ±r.

Veri Ã¶lÃ§eklendirme, genellikle iki temel yÃ¶ntemle yapÄ±lÄ±r:

**Min-Max Ã–lÃ§ekleme (Normalization):** Bu yÃ¶ntemde, veriler belirli bir aralÄ±ÄŸa (genellikle 0 ile 1 arasÄ±na) dÃ¶nÃ¼ÅŸtÃ¼rÃ¼lÃ¼r. Her bir deÄŸer, orijinal deÄŸerin minimum ve maksimum deÄŸerleri arasÄ±nda nerede olduÄŸunu yansÄ±tan bir skalama iÅŸlemi uygulanÄ±r.

**StandartlaÅŸtÄ±rma (Standardization):** Bu yÃ¶ntemde, her Ã¶zellik, ortalama deÄŸerinden Ã§Ä±karÄ±lÄ±r ve ardÄ±ndan standart sapmaya bÃ¶lÃ¼nÃ¼r. Bu, her Ã¶zelliÄŸin ortalama 0 ve standart sapma 1'e yakÄ±n bir deÄŸere sahip olmasÄ±nÄ± saÄŸlar.





```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
X_test
```

**MODELS**

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_squared_error,accuracy_score
from sklearn.model_selection import cross_val_score

```

```python
# Create and train the model
lr = LinearRegression()
lr.fit(X_train,y_train)

dt = DecisionTreeRegressor()
dt.fit(X_train,y_train)

rf = RandomForestRegressor()
rf.fit(X_train,y_train)

gb = GradientBoostingRegressor()
gb.fit(X_train,y_train)

xgb = XGBRegressor()
xgb.fit(X_train,y_train)
```

Cross Validation for model performance

```python
models = [lr, dt, rf, gb, xgb]

compare_df = pd.DataFrame(columns=['Model', 'Cross-Validated RMSE'])

# Modelleri karÅŸÄ±laÅŸtÄ±rma
for model in models:
    cv_results = cross_val_score(model, X_train, y_train, cv=10, scoring='neg_mean_squared_error')
    cv_rmse = np.sqrt(abs(cv_results.mean()))
    model_name = type(model).__name__  # Model adÄ±nÄ± alÄ±nmasÄ±
    compare_df = compare_df.append({'Model': model_name, 'Cross-Validated RMSE': cv_rmse}, ignore_index=True)


print(compare_df)
```

 En dÃ¼ÅŸÃ¼k RMSE deÄŸeri daha iyi bir model performansÄ±nÄ± gÃ¶sterir. Yani, en kÃ¼Ã§Ã¼k RMSE deÄŸeri olan model daha iyi tahminler yapmÄ±ÅŸ olur.Bu durumda **"GradientBoostingRegressor"** modeli, en dÃ¼ÅŸÃ¼k Ã§apraz doÄŸrulama RMSE deÄŸeri olan 4471.77 ile daha iyi tahmin yapmÄ±ÅŸ gibi gÃ¶rÃ¼nÃ¼yor.

**Gradient Boosting Regressor (Gradient ArtÄ±rma Regresyonu):**

**Hyper-parameter Optimization- Improving the Model**


> Gradient Boosting Regressor modelinde en yaygÄ±n kullanÄ±lan hiperparametreler

*   **n_estimators:** OluÅŸturulan aÄŸaÃ§ sayÄ±sÄ±.
*   **learning_rate:** Her aÄŸacÄ±n katkÄ±sÄ±nÄ±n kÃ¼Ã§Ã¼ltÃ¼lme hÄ±zÄ±.
*  **max_depth:** Her bir aÄŸacÄ±n maksimum derinliÄŸi.
*  **min_samples_split:** Bir dÃ¼ÄŸÃ¼mÃ¼ bÃ¶lmek iÃ§in gerekli minimum Ã¶rnek sayÄ±sÄ±.
* **min_samples_leaf:** Bir yapraÄŸÄ± oluÅŸturmak iÃ§in gereken minimum Ã¶rnek sayÄ±sÄ±.
*  **subsample:** Her aÄŸaÃ§ iÃ§in kullanÄ±lacak alt Ã¶rneklem oranÄ±.


Grid Search ve Randomized Search ile hiperparametre optimizasyonu :

**Grid Search (Kafes Arama):** Belirli bir hiperparametre uzayÄ±nda tÃ¼m olasÄ± kombinasyonlarÄ± deneyerek en iyi performansÄ± saÄŸlayan hiperparametreleri bulur. Bu yÃ¶ntem kapsamlÄ± bir arama yapar ancak bÃ¼yÃ¼k hiperparametre uzaylarÄ± durumunda maliyetli olabilir.

**Randomized Search (Rastgele Arama):** Belirli bir hiperparametre uzayÄ±ndan rastgele Ã¶rneklem alarak arama yapar. Bu yÃ¶ntem, Grid Search'a gÃ¶re daha hÄ±zlÄ±dÄ±r ve bÃ¼yÃ¼k hiperparametre uzaylarÄ±nda kullanÄ±ÅŸlÄ±dÄ±r.

```python
from sklearn.model_selection import GridSearchCV

# Optimizasyon iÃ§in hiperparametre deÄŸerlerini
param_grid = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 4, 5],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'subsample': [0.8, 0.9, 1.0]
}

# Grid Search iÅŸlemi
grid_search = GridSearchCV(gb, param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_
best_model = grid_search.best_estimator_
```

```python
# Best Params
print(best_params)
```

```python
# # Predict on the test set using the best model(optimized model)
y_pred = best_model.predict(X_test)

```

Evaluation Model

```python
#RMSE Calculating
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print(f"Root Mean Squared Error: {rmse}")
print(f"Mean Squared Error: {mse}")
```

```python
importances = pd.Series(best_model.feature_importances_, index = X.columns)
sort = importances.sort_values()
sort.plot(kind = 'barh', color = 'lightgreen')
plt.show()
print('R2', best_model.score(X_test, y_test))
```

Feature Selection

```python
from sklearn.feature_selection import RFE
n = [1, 2, 3, 4, 5, 6, 7, 8]

for n in n:
    gbr2 = RFE(estimator = best_model, n_features_to_select = n, verbose = 0)
    gbr2.fit(X_train, y_train)
    print(f'R2 for {n} features', gbr2.score(X_test, y_test))
    print(f'RMSE', np.sqrt(mean_squared_error(y_test, gbr2.predict(X_test))))
```

RMSE ve en yÃ¼ksek test seti PuanlarÄ± 4 ve 5'dÄ±r. BakalÄ±m bunlar hangi Ã¶zellikler.
