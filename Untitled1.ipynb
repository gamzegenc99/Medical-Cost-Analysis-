There are outlier values on the top, we need to suppress them as below:


```python

Q1 = df['bmi'].quantile(0.25)
Q3 = df['bmi'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df['bmi'] = df['bmi'].apply(lambda x: upper_bound if x > upper_bound else (lower_bound if x < lower_bound else x))

sns.boxplot(df['bmi'])
plt.title('Box plot of BMI')
plt.show()
```


**Correlation**
> Değişkenler arasındaki korelasyon, bir değişkenin değeri diğer değişkenin değeriyle nasıl bir ilişki içerisinde olduğunu belirtir

> Korelasyon, -1 ile 1 arasında bir değer alır:

Pozitif korelasyon (0 ile 1 arası): İki değişken arasında pozitif bir ilişki olduğunu gösterir. Bir değişkenin değeri artarken diğer değişkenin değeri de artar. Değer 1'e yaklaştıkça ilişki daha güçlüdür.

Negatif korelasyon (-1 ile 0 arası): İki değişken arasında negatif bir ilişki olduğunu gösterir. Bir değişkenin değeri artarken diğer değişkenin değeri azalır. Değer -1'e yaklaştıkça ilişki daha güçlüdür.

Korelasyon yok (0): İki değişken arasında herhangi bir belirgin ilişki bulunmamaktadır.

**Data splitting**


```python
X = df.drop('charges', axis = 1)
y = df['charges']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.20, random_state = 0)
```

**Scaling**

>  Veri ölçeklendirme, farklı özelliklerin veya değişkenlerin farklı aralıklara sahip olduğu durumlarda bu değerleri belirli bir ölçekte birleştirmeyi ifade eder. Bu, modelin dengeli ve daha iyi sonuçlar üretmesine yardımcı olur.


> Veri ölçeklendirme, genellikle şu iki ana nedenden dolayı yapılır:

**Makine Öğrenimi Algoritmalarının Performansı:** Birçok makine öğrenimi algoritması, özellikler arasındaki büyüklük farklarını dikkate almaz. Bu nedenle, ölçeklenmemiş verilerle çalışırken, büyük değerlere sahip özellikler diğerlerini baskılayabilir ve modelin yanlış sonuçlar üretmesine neden olabilir.

**Optimizasyon ve Hızlandırma:** Bazı algoritmalar, verinin ölçeklenmiş olmasının modelin eğitimini hızlandırabileceği veya daha iyi sonuçlar verebileceği durumlar vardır.

Veri ölçeklendirme, genellikle iki temel yöntemle yapılır:

**Min-Max Ölçekleme (Normalization):** Bu yöntemde, veriler belirli bir aralığa (genellikle 0 ile 1 arasına) dönüştürülür. Her bir değer, orijinal değerin minimum ve maksimum değerleri arasında nerede olduğunu yansıtan bir skalama işlemi uygulanır.

**Standartlaştırma (Standardization):** Bu yöntemde, her özellik, ortalama değerinden çıkarılır ve ardından standart sapmaya bölünür. Bu, her özelliğin ortalama 0 ve standart sapma 1'e yakın bir değere sahip olmasını sağlar.





```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
X_test
```

**MODELS**

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_squared_error,accuracy_score
from sklearn.model_selection import cross_val_score

```

```python
# Create and train the model
lr = LinearRegression()
lr.fit(X_train,y_train)

dt = DecisionTreeRegressor()
dt.fit(X_train,y_train)

rf = RandomForestRegressor()
rf.fit(X_train,y_train)

gb = GradientBoostingRegressor()
gb.fit(X_train,y_train)

xgb = XGBRegressor()
xgb.fit(X_train,y_train)
```

Cross Validation for model performance

```python
models = [lr, dt, rf, gb, xgb]

compare_df = pd.DataFrame(columns=['Model', 'Cross-Validated RMSE'])

# Modelleri karşılaştırma
for model in models:
    cv_results = cross_val_score(model, X_train, y_train, cv=10, scoring='neg_mean_squared_error')
    cv_rmse = np.sqrt(abs(cv_results.mean()))
    model_name = type(model).__name__  # Model adını alınması
    compare_df = compare_df.append({'Model': model_name, 'Cross-Validated RMSE': cv_rmse}, ignore_index=True)


print(compare_df)
```

 En düşük RMSE değeri daha iyi bir model performansını gösterir. Yani, en küçük RMSE değeri olan model daha iyi tahminler yapmış olur.Bu durumda **"GradientBoostingRegressor"** modeli, en düşük çapraz doğrulama RMSE değeri olan 4471.77 ile daha iyi tahmin yapmış gibi görünüyor.

**Gradient Boosting Regressor (Gradient Artırma Regresyonu):**

Gradient artırma regresyonu, zayıf tahmincileri (genellikle karar ağaçları) birleştirerek daha güçlü bir tahminci oluşturur.
Her aşamada, önceki aşamaların hatalarını düzeltmeye çalışır.
Aşırı öğrenmeye karşı dirençlidir ve yüksek tahmin performansı sunar.

**Hyper-parameter Optimization- Improving the Model**


> Gradient Boosting Regressor modelinde en yaygın kullanılan hiperparametreler

*   **n_estimators:** Oluşturulan ağaç sayısı.
*   **learning_rate:** Her ağacın katkısının küçültülme hızı.
*  **max_depth:** Her bir ağacın maksimum derinliği.
*  **min_samples_split:** Bir düğümü bölmek için gerekli minimum örnek sayısı.
* **min_samples_leaf:** Bir yaprağı oluşturmak için gereken minimum örnek sayısı.
*  **subsample:** Her ağaç için kullanılacak alt örneklem oranı.


Grid Search ve Randomized Search ile hiperparametre optimizasyonu :

**Grid Search (Kafes Arama):** Belirli bir hiperparametre uzayında tüm olası kombinasyonları deneyerek en iyi performansı sağlayan hiperparametreleri bulur. Bu yöntem kapsamlı bir arama yapar ancak büyük hiperparametre uzayları durumunda maliyetli olabilir.

**Randomized Search (Rastgele Arama):** Belirli bir hiperparametre uzayından rastgele örneklem alarak arama yapar. Bu yöntem, Grid Search'a göre daha hızlıdır ve büyük hiperparametre uzaylarında kullanışlıdır.

```python
from sklearn.model_selection import GridSearchCV

# Optimizasyon için hiperparametre değerlerini
param_grid = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 4, 5],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'subsample': [0.8, 0.9, 1.0]
}

# Grid Search işlemi
grid_search = GridSearchCV(gb, param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_
best_model = grid_search.best_estimator_
```

```python
# Best Params
print(best_params)
```

```python
# # Predict on the test set using the best model(optimized model)
y_pred = best_model.predict(X_test)

```

Evaluation Model

```python
#RMSE Calculating
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print(f"Root Mean Squared Error: {rmse}")
print(f"Mean Squared Error: {mse}")
```

```python
importances = pd.Series(best_model.feature_importances_, index = X.columns)
sort = importances.sort_values()
sort.plot(kind = 'barh', color = 'lightgreen')
plt.show()
print('R2', best_model.score(X_test, y_test))
```

Feature Selection

```python
from sklearn.feature_selection import RFE
n = [1, 2, 3, 4, 5, 6, 7, 8]

for n in n:
    gbr2 = RFE(estimator = best_model, n_features_to_select = n, verbose = 0)
    gbr2.fit(X_train, y_train)
    print(f'R2 for {n} features', gbr2.score(X_test, y_test))
    print(f'RMSE', np.sqrt(mean_squared_error(y_test, gbr2.predict(X_test))))
```

RMSE ve en yüksek test seti Puanları 4 ve 5'dır. Bakalım bunlar hangi özellikler.

```python
n = [4, 5]

for n in n:
    gbr3 = RFE(estimator = best_model, n_features_to_select = n, verbose = 0)
    gbr3.fit(X_train, y_train)
    print(f'{n} Features', X.loc[:, gbr3.support_].columns)
```
